```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hb=pd.read_csv('headbrain.csv')
```

```python
hb.head()
```

```python
hb.nunique()
```

```python
hb.isnull().sum() 
```

```python
sns.boxplot(hb['Head Size(cm^3)'])
plt.title('Box plot')
```

```python
sns.violinplot(hb['Head Size(cm^3)'])
plt.title('Violin plot')
```

```python
sns.distplot(hb['Head Size(cm^3)'],bins=10, kde=False,hist_kws=dict(edgecolor="k", linewidth=2),hist=True)
plt.title('Histogram')
```

```python
plt.figure(figsize=(10,12))
plt.subplot(2, 1, 1)
sns.violinplot(hb['Brain Weight(grams)'])
plt.title('Violin plot')

plt.subplot(2,1,2)
sns.distplot(hb['Brain Weight(grams)'],bins=10, kde=False,hist_kws=dict(edgecolor="k", linewidth=2),hist=True)
plt.title('Histogram')
```

```python
sns.boxplot(hb['Brain Weight(grams)'])
plt.title('Box plot')
```

```python
sns.scatterplot(y='Brain Weight(grams)',x='Head Size(cm^3)',data=hb)
plt.show()
```

```python
sns.jointplot(x="Brain Weight(grams)", y="Head Size(cm^3)", data=hb,kind="hex",height=8,color="g") 
plt.show()
```
